## ⚒️Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Attrition'] = le.fit_transform(df['Attrition'])

le1 = LabelEncoder()
df['Gender'] = le1.fit_transform(df['Gender'])

le2 = LabelEncoder()
df['JobRole'] = le2.fit_transform(df['JobRole'])

le3 = LabelEncoder()
df['OverTime'] = le3.fit_transform(df['OverTime'])

le4 = LabelEncoder()
df['Over18'] = le4.fit_transform(df['Over18'])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe1 = OneHotEncoder(sparse_output=False)
ohe1.fit(df[['BusinessTravel']])
col = ohe1.get_feature_names_out(['BusinessTravel'])
df1 = pd.DataFrame(ohe1.transform(df[['BusinessTravel']]),columns=col)

ohe2 = OneHotEncoder(sparse_output=False)
ohe2.fit(df[['Department']])
col1 = ohe2.get_feature_names_out(['Department'])
df2 = pd.DataFrame(ohe2.transform(df[['Department']]),columns=col1)

ohe3 = OneHotEncoder(sparse_output=False)
ohe3.fit(df[['EducationField']])
col2 = ohe3.get_feature_names_out(['EducationField'])
df3 = pd.DataFrame(ohe3.transform(df[['EducationField']]),columns=col2)

ohe4 = OneHotEncoder(sparse_output=False)
ohe4.fit(df[['MaritalStatus']])
col3 = ohe4.get_feature_names_out(['MaritalStatus'])
df4 = pd.DataFrame(ohe4.transform(df[['MaritalStatus']]),columns=col3)
```

```python
df = pd.concat([df,df1,df2,df3,df4],axis=1)
df.drop(['BusinessTravel','Department','EducationField','MaritalStatus'],axis=1,inplace=True)
```

## ✂️Split the Data into Training set and Testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

**Standardise the data**

```python
from sklearn.preprocessing import StandardScaler


sd = StandardScaler()

sd.fit(x)
xstand = sd.transform(x)
```

## 📈Model Building

### Logistic Regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
import numpy as np

param_grid = {
    'C': np.logspace(-3, 3, 7), 
    'penalty': ['l1', 'l2'],    
    'solver': ['liblinear', 'saga']
}

lr = LogisticRegression(max_iter=1000)

LR = GridSearchCV(
    estimator=lr,
    param_grid=param_grid,
    scoring='accuracy',  
    cv=5,                 
    verbose=2,
    n_jobs=-1
)

LR.fit(x_train,y_train)
```

```python
print("Best Parameters:", LR.best_params_)
print("Best Cross-validation Score:", LR.best_score_)
```

```python
Model = LR.best_estimator_
y_pred = Model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Attrition", "Attrition"])
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state=42)
RF.fit(x_train, y_train)

RF_pred = RF.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(accuracy_score(y_test,RF_pred))
print(confusion_matrix(y_test,RF_pred))
print(classification_report(y_test,RF_pred))
```
